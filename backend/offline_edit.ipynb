# Edit Personas

```python
df = pd.read_pickle("/Users/nlotan/code/university/personal_llm_vc/backend/app/data/persona_details_v3.pkl")
```

```python
df = df[~df['description'].isna()]
```

```python
import re

def split_description(description):
#    if not isinstance(description, str):
#        return description
    return re.split(r'example\s+accounts\s+(followed|include)\s*:', description, flags=re.IGNORECASE)[0].strip()
```

```python
# edit the decription column of all rows to remove any text after example accounts followed: including the text "Example accounts include:"
df['description'] = df['description'].progress_apply(split_description)
```
